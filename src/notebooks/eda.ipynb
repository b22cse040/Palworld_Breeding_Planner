```python
import pandas as pd 

df = pd.read_json('../../data/pals.json')
df.head()
```

```python
for i in range(0, 3):
    print(f"Pal Idx {i} \nName: {df['name'][i]}\nType -> {df['suitability'][i]}\n")
```

```python
abilities : set[str] = set()

for idx, row in df.iterrows():
    types : list[dict[str, str]] = df['suitability'][idx]
    [abilities.add(dicct['type']) for dicct in types]

print(abilities)
```

```python
abilities = set()

for types in df['suitability']:
    for d in types:
        abilities.add(d['type'])

abilities = sorted(abilities)
```

```python
for ability in abilities:
    df[ability] = 0
```

```python
df.head()
```

```python
df['suitability'][0]
```

```python
for idx, row in df.iterrows():
    suitability : list[dict[str, str]] = row['suitability']
    for entry in suitability:
        ability = entry['type']
        df.at[idx, ability] = entry['level']

df.head()
```

```python
df = df.drop(['image', 'suitability', 'imageWiki', 'drops', 'asset', 'genus', 'size', 'maps', 'price'], axis = 1)
df.head()
```

### Breeding.json

```python
import json
import pandas as pd

with open("../../data/breeding.json", "r") as f:
    breed = json.load(f)

print(json.dumps(breed, indent=2))
```

```python
rows = []

for child_id, parent_pairs in breed.items():
    for p1, p2 in parent_pairs:
        rows.append({
            "child_id": child_id,
            "first_parent_id": p1,
            "second_parent_id": p2
        })

breed_df = pd.DataFrame(rows)
breed_df.head()
```

```python
from typing import Dict, Tuple
breed_df["p1"] = breed_df[["first_parent_id", "second_parent_id"]].min(axis=1)
breed_df["p2"] = breed_df[["first_parent_id", "second_parent_id"]].max(axis=1)

breed_df = breed_df.drop(["first_parent_id", "second_parent_id"], axis = 1)
breed_df.head()
```

```python
breed_df.to_parquet("../data/breeding.parquet")
```
